# EDA - Student Mental Health

Dataset: https://www.kaggle.com/datasets/shariful07/student-mental-health


```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def find_csv():
    if os.environ.get('DATA_PATH') and os.path.exists(os.environ['DATA_PATH']):
        return os.environ['DATA_PATH']
    raw_dir = os.path.join('data','raw')
    if os.path.isdir(raw_dir):
        for f in os.listdir(raw_dir):
            if f.lower().endswith('.csv'):
                return os.path.join(raw_dir,f)
    raise FileNotFoundError('Place CSV under data/raw/ or set DATA_PATH')

path = find_csv()
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
num_cols = df.select_dtypes(include=['number','bool']).columns
corr = df[num_cols].corr(numeric_only=True) if len(num_cols) else None
corr
```

```python
plt.figure(figsize=(10,8))
if corr is not None and corr.shape[0] >= 2:
    sns.heatmap(corr, cmap='coolwarm', center=0)
    plt.title('Correlation Heatmap')
plt.show()
```

```python
hist_col = 'Age' if 'Age' in df.columns else (num_cols[0] if len(num_cols) else None)
plt.figure(figsize=(8,5))
if hist_col is not None:
    sns.histplot(df[hist_col].dropna(), kde=True)
    plt.title(f'Histogram: {hist_col}')
plt.show()
```

```python
target = 'Depression' if 'Depression' in df.columns else df.columns[-1]
plt.figure(figsize=(8,5))
sns.countplot(x=df[target].astype(str))
plt.title(f'Countplot: {target}')
plt.xticks(rotation=30, ha='right')
plt.show()
```
